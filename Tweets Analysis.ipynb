```python
import pandas as pd
import json
import wordcloud as wc
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
list_tweets = []
with open('C:\\Users\\Lenovo\\Desktop\\TestFiles\\twitter_data.txt','r') as td:
    for lin in td:
        try:
            lin = lin.strip()
            line = json.loads(lin)
            if type(line) is dict:
                list_tweets.append(line)
        except:
            pass
```

```python
total_tweets = len(list_tweets)
```

```python
df = pd.DataFrame(list_tweets)
```

```python
df.drop(['contributors','coordinates'],axis=1,inplace=True)
```
